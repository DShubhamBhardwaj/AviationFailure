```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

READING DATA SET

```python
raw_data= pd.read_csv('CMaps/train_FD001.txt', sep=' ', header = None)
raw_data
```

```python
raw_data = raw_data.drop([26,27], axis='columns')
raw_data
```

```python
raw_data.columns = ['ID', 'Cycle', 'OpSet1', 'OpSet2', 'OpSet3', 'SensorMeasure1', 'SensorMeasure2', 'SensorMeasure3', 'SensorMeasure4', 'SensorMeasure5', 'SensorMeasure6', 'SensorMeasure7', 'SensorMeasure8', 'SensorMeasure9', 'SensorMeasure10', 'SensorMeasure11', 'SensorMeasure12', 'SensorMeasure13', 'SensorMeasure14', 'SensorMeasure15', 'SensorMeasure16', 'SensorMeasure17', 'SensorMeasure18', 'SensorMeasure19', 'SensorMeasure20', 'SensorMeasure21']
raw_data
```

RUL = EOL - Cycle

- RUL = Remaining Useful Life
- EOL = End of Life in cycles

```python
EOL=[]
for sl_no in raw_data['ID']:
       EOL.append( ((raw_data[raw_data['ID'] == sl_no]["Cycle"]).values)[-1]) 
        
raw_data["EOL"]=EOL
raw_data
```

<h1>Defining the labels</h1>
Here we will define the engine's condition with Life Ratio (LR), which is the ratio between Current Cycle and the End cycle/ End of Life(EOL). If LR=0, that means the component has just started its degradation and LR=1 means, it is completely degraded.
<ul>
    <li>if LR Less than or equal to 0.6 - Good Condition</li>
    <li>if LR is between 0.6  and 0.8 - Moderate Condition</li>
    <li>if LR greater than 0.8 - Warning Condition</li>
</ul>    

```python
raw_data["LR"] = raw_data["Cycle"].div(raw_data["EOL"])
raw_data
```

```python
labels=[]
for i in range (0,len(raw_data)):
    if np.array(raw_data["LR"])[i] <= 0.6:
        labels.append(0)
    elif np.array(raw_data["LR"])[i] <= 0.8 :
        labels.append(1)   
    else :
        labels.append(2)  
        
raw_data["labels"]=labels
raw_data  = raw_data.drop(columns=['LR'])   
raw_data
```

```python
raw_data['RUL'] = raw_data.groupby(['ID'])['Cycle'].transform(max)-raw_data['Cycle']
df_train=raw_data
```

<h1>DATA SET PRE PROCESSING !</h1>

```python
fig,ax=plt.subplots(1,3,figsize=(30,7),sharex='all')
for i in range(0,3):
    df_u1=df_train.query('ID==3')
    ax[i].plot(df_u1.RUL.values, df_u1['OpSet'+str(i+1)])
    ax[i].set_title('OpSet'+str(i+1))
```

```python
fig,ax=plt.subplots(7,3,figsize=(30,40),sharex=True)
df_u1=df_train.query('ID==4')
c=0
for i in range(0,7):
    for j in range(0,3):
        ax[i,j].plot(df_u1.RUL.values, df_u1['SensorMeasure'+str(c+1)])
        ax[i,j].set_title('SensorMeasure'+str(c+1))
        ax[i,j].axvline(0,c='r')
        c+=1
plt.suptitle('Sensor Traces: Dataset 1',fontsize=50)
plt.show()
```

```python
df_train = df_train.drop(columns=delete_columns)
df_train =df_train.drop(columns=['RUL'])
```
